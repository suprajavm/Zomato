TOP RATED CUISINES IN INDIA

```python
import pandas as pd
from sqlalchemy import create_engine

username="root"
password=123456
database="zomato"
port=3306
host="localhost"

engine=create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

query="""With cte as
(select m.cuisines,m.country_code,m.rating_text,c.country,m.city,m.aggregate_rating from mytable m left join country_code c on m.country_code=c.country_code)
select cuisines,city,country,AVG(aggregate_rating) AS avg_rating from cte where country="India" GROUP BY 
    cuisines, city, country
limit 20;""" 

a=pd.read_sql(query,engine)

sns.barplot(y="cuisines",x="avg_rating",data=a)


plt.show()
```

Relationship Between Price Range and Ratings

```python
query="""SELECT 
    price_range,
    COUNT(*) AS num_restaurants,
    AVG(aggregate_rating) AS avg_rating,
    SUM(votes) AS total_votes
FROM 
    mytable
GROUP BY 
    price_range
ORDER BY 
    price_range;"""

b=pd.read_sql(query,engine)
x=np.array(b["price_range"])
y=np.array(b["avg_rating"])
plt.bar(x,y)
plt.show()
```

Online Delivery Impact on Ratings and Popularity

```python

query="select has_online_delivery,count(*) as no_of_restauarnts,avg(aggregate_rating) as avg_rating, sum(votes) as total_votes from mytable group by has_online_delivery order by total_votes DESC;"
c=pd.read_sql(query,engine)

plt.subplot(2,2,1)
sns.barplot(x="has_online_delivery",y="avg_rating",data=c)
plt.subplot(2,2,2)
sns.barplot(x="has_online_delivery",y="total_votes",data=c) 
plt.tight_layout() 
plt.show()
```

Customer Voting Patterns for Different Cuisines

```python
query="select cuisines, sum(votes) as total_votes from mytable group by cuisines order by total_votes desc limit 20; "

e=pd.read_sql(query,engine)

sns.barplot(x="total_votes",y="cuisines",data=e)
plt.show()

```

Currency and Price Range Analysis of India

```python
query="""with cte as(
select m.price_range,m.city,m.country_code,c.country,m.Currency from mytable m left join country_code c on m.country_code=c.country_code)
select avg(price_range) as average_price_range,city,country,Currency from cte where country="India" group by country,city,currency;
"""

f=pd.read_sql(query,engine)

plt.figure(figsize=(12, 6))

sns.barplot(y="city",x="average_price_range",data=f)
plt.tight_layout()
plt.show()



```

Rating Color Distribution Analysis of New Delhi

```python
query="""with cte as(
select m.city,m.country_code,m.rating_color,c.country from mytable m left join country_code c on m.country_code=c.country_code),
 cte1 as(select city,country,count(*) as total_hotels from cte group by city,country)select cte.city,cte.country,cte.rating_color,(count(*)*100)/cte1.total_hotels as percentage from cte left join cte1 on cte.city=cte1.city and cte.country=cte1.country where cte.city="New Delhi" group by cte.city,cte.country,cte.rating_color;"""

f=pd.read_sql(query,engine)

sns.barplot(x="rating_color",y="percentage",data=f)
plt.show()
```

Common Cuisines Offered in High-Density Restaurant Areas of New Delhi

```python
query="""with cte as(
select m.cuisines,m.city,m.country_code,c.country ,m.aggregate_rating from mytable m left join country_code c on m.country_code=c.country_code)
select cuisines,city,country,avg(aggregate_rating) as average_rating ,count(*) as no_of_restuarnts from cte where city="New Delhi" group by cuisines,city,country limit 20;"""

g=pd.read_sql(query,engine) 

sns.barplot(x="average_rating",y="cuisines",data=g)
plt.show()

```
